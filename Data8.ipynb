```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,PowerTransformer,QuantileTransformer
import matplotlib.pyplot as plt
import numpy as np
sns.set_theme(style="darkgrid")
```

```python
data=pd.read_csv("IRIS.csv")
iris=data
iris
```

```python
fig,ax=plt.subplots(ncols=2,nrows=2)
sns.histplot(data=iris,x='sepal_length',bins=100,kde=True,hue='species',ax=ax[0,0]).set(title='sepal_length')
# plt.show()
sns.histplot(data=iris,x='sepal_width',bins=100,kde=True,hue='species',ax=ax[0,1]).set(title='sepal_width')
sns.histplot(x = 'petal_length' , data = iris , bins = 100, kde=True, hue='species', ax=ax[1, 0]).set(title='petal_length')
sns.histplot(x = 'petal_width' , data = iris , bins = 100, kde=True, hue='species', ax=ax[1, 1]).set(title='petal_width')
plt.show()
plt.tight_layout()
fig.set_figwidth(7)
fig.set_figheight(7)
```

```python
fig,ax=plt.subplots(ncols=2,nrows=2)
sns.boxplot(data=iris,x='sepal_length',hue='species',ax=ax[0,0],legend=True).set(title='sepal_length')
# plt.show()
sns.boxplot(data=iris,x='sepal_width',hue='species',ax=ax[0,1],legend=True).set(title='sepal_width')
sns.boxplot(x = 'petal_length' , data = iris ,  hue='species', ax=ax[1, 0],legend=True).set(title='petal_length')
sns.boxplot(x = 'petal_width' , data = iris ,  hue='species', ax=ax[1, 1],legend=True).set(title='petal_width')
# plt.show()
plt.tight_layout()
fig.set_figwidth(10)
fig.set_figheight(10)
```
