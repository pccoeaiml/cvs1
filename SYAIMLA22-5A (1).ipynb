```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
df=pd.read_csv("wine.csv",usecols=[0,1,2])
df.columns=["Wine","Alcohol","Malic.acid"]
a=MinMaxScaler()
b=a.fit_transform(df[['Alcohol','Malic.acid']])
print(b)
```
