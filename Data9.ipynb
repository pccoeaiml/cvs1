```python
import seaborn as sns
```

```python
import pandas as pd
from scipy.stats import zscore
from sklearn.preprocessing import MinMaxScaler,PowerTransformer,QuantileTransformer
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
sns.set_theme(style="darkgrid")
sns.set_color_codes("pastel")
```

```python
df=pd.read_csv("diabetes9.csv")
df
```

```python
df.head()
```

```python
sns.lineplot(x='Age',y='Glucose',data=df)
plt.show()
```

```python
mean=np.mean(df['Glucose'])
mean
```

```python
sns.boxplot(x='Glucose',data=df)
```

```python
sns.scatterplot(x='Glucose', y='BloodPressure', data=df, hue='Outcome', palette='viridis')

# Add a title
plt.title('Scatter Plot: Glucose vs BloodPressure')

# Show the plot
plt.show()
```

```python
sns.barplot(x='Glucose', y='BloodPressure', data=df, hue='Outcome', palette='viridis')

# Add a title
plt.title('Scatter Plot: Glucose vs BloodPressure')

# Show the plot
plt.show()
```

```python
x=df['Glucose']
y=df['BloodPressure']
plt.plot(x,y)
plt.xlabel('Glucose')
plt.ylabel('bP')
plt.show()
```

```python
sns.lineplot(x='Age', y='Glucose', data=df)
plt.title('Line Plot: Glucose levels over Age')
plt.show()
```

```python
sns.scatterplot(data=df,x='Glucose',y='BMI',hue="Outcome")
plt.show()
```

```python
sns.displot(data=df,x='BloodPressure',y='Glucose',hue='BMI')
```

```python
sns.color_palette("Greens")
sns.scatterplot(data=df,x='Glucose',y='BMI',hue='Age')
plt.show()
plt.figure()
sns.lineplot(data=df,x='Age',y='BloodPressure',hue='BloodPressure')
plt.show()
```

```python
sns.barplot(x='Age',y='Insulin',data=df,hue='Age',legend=True)
```

```python
sns.kdeplot(data=df,x='Age',y='Glucose')
```

```python
sns.regplot(data=df,x='Age',y='Outcome',ci=None)
plt.show()
```

```python

plt.figure(figsize=(10, 8))

# Create a bubble plot
sns.scatterplot(x='Age', y='Glucose', size='BMI', sizes=(20, 200), data=df, hue='Outcome', palette='viridis', alpha=0.7)

# Add title and labels
plt.title('Bubble Plot: Age vs Glucose (BMI as size)')
plt.xlabel('Age')
plt.ylabel('Glucose')

# Add a legend
plt.legend(title='Outcome')

# Show the plot
plt.show()
```

```python
sns.scatterplot(data=df,x='Age',y='Glucose',hue='BMI')
plt.title("Age vs Glucose ")
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add a title
plt.title('Correlation Matrix')

# Show the plot
plt.show()
```
